#  <div style="text-align: center"> Interpolare cu functii Spline </div>

##  <div style="text-align: center"> De ce? </div>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Pana acum am analizat 2 metode de interpolare ce aveau ca rezultat un singur polinom ce trecea prin toate punctele date. Ce probleme apareau?
- erori mari in capetele domeniului de interpolare la Lagrange, din cauza gradului mare al polinomului
- complexitate de calcul exponentiala la Neville, daca nu se foloseste memorizarea: $O(2^n)$

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; In aparenta, ar fi necesar un compromis perpetuu intre complexitate si erori. Cu toate acestea, exista moduri in care putem obtine o interpolare buna, fara a face compromisuri: **functiile Spline.** Pentru a scapa de problema polinoamelor de grad mare, ce pot avea oscilatii greu de controlat, se vor interpola cate **2 puncte consecutive**, functiile rezultate fiind alese astfel incat sa rezulte o curba cat mai **neteda**.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Se cauta, deci, o functie $S$ definita pe ramuri, fiecare dintre acestea fiind o alta functie, numita **functie de imbinare** $s_0, s_1, \dots, s_{n -1}$ care sa asigure interpolari intre cate 2 puncte consecutive ale functiei date. Putem defini $S(x)$ astfel:

$$
S(x) = \begin{cases} s_0(x), & x \in [x_0, x_1] \\ s_1(x), & x \in [x_1, x_2] \\ . \\ . \\ . \\ s_{n - 1}(x), & x \in [x_{n - 1}, x_n] \end{cases}
$$

##  <div style="text-align: center"> 1. Interpolare cu functii de clasa $C^0$ </div>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Intr-o prima faza ne punem problema gasirii unei functii $S$ **continue** care sa interpoleze toata functia data si sa treaca prin punctele-suport. Functiile de imbinare $s_0, s_1, \dots, s_n$ vor fi niste drepte, deci vor fi polinoame de grad 1:

$$s_i(x) = a_ix + b_i; \  a_i, b_i \in \mathbb{R}, i = \overline{0,n - 1}$$

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Dupa cum se observa, sunt $2n$ necunoscute, $a_i$ si $b_i$ cu $i = \overline{0,n - 1}$. Deci, vom avea nevoie de $2n$ ecuatii pentru a le determina. Aceste ecuatii rezulta din 2 tipuri de conditii fundamentale pentru interpolarile cu functii Spline:

### Conditii de interpolare

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Aceste conditii se asigura ca $S$ trece prin toate punctele-suport:

$s_i(x_i) = f(x_i), \  i = \overline{0,n - 1} \Rightarrow n$ ecuatii

$s_{n - 1}(x_n) = f(x_n)$

### Conditii de racordare

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Functia $S \in C^0([x_0, x_n])$, deci este continua, iar conditiile de racordare vizeaza aceasta proprietate:

$s_i(x_{i + 1}) = s_{i + 1}(x_{i + 1}), \  i = \overline{0,n - 1}  \Rightarrow$ **alte** $\textbf{n - 1}$ **ecuatii**

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; In total, din aceste 2 conditii rezulta cele $2n$ ecuatii necesare. Sistemul va fi de forma:

$a_ix_i + b_i = f(x_i), \ i = \overline{0,n - 1}$

$a_{n - 1}x_n + b_n = f(x_n)$

$(a_i - a_{i + 1})x_{i + 1} + (b_i - b_{i + 1}) = 0, \ i = \overline{0,n - 1}$, trecand membrul drept din conditiile de racordare in partea stanga

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Dupa rezolvarea acestui sistem, va rezulta o interpolare de forma:

![ref1](Spline_C0.png 'Graficul interpolarii cu functii Spline continue, in diverse puncte')

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Observati natura erorilor in fiecare caz! Asadar, pentru o eroare de ordinul $10^{-3}$ este nevoie in acest caz de de 3 ori mai multe puncte-suport. Pe langa posibilitatea de a nu dispune de acele puncte suplimentare, calcularea Spline-urilor in acestea ar face ca numarul de operatii sa creasca de $9$ ori (folosind metode iterative, complexitatea de rezolvare a unui sistem este $O(n^2)$).

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Deci Spline-urile de clasa $C^0$ nu sunt prea eficiente daca numarul de puncte din suportul interpolarii este mic. Daca ne uitam la functia data in graficele de mai sus, aceasta nu pare doar continua, ci pare ca are si **derivata continua**. Ce ar fi sa imitam acest comportament?

##  <div style="text-align: center"> 2. Interpolare cu functii de clasa $C^1$

### Ce se schimba?

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Care sunt diferentele dintre o functie de clasa $C^0$ si una de clasa $C^1$? Se pot observa in graficele de mai jos. In cel de sus avem [Fractalul lui Weierstrass](https://en.wikipedia.org/wiki/Weierstrass_function), o functie continua, pe $\mathbb{R}$, dar nederivabila in niciun punct, iar jos este rezultatul **integrarii** functiei care genereaza acest fractal, adica o functie de clasa $C^1$, dar care nu are derivata de ordin 2 in niciun punct. Observati ca aceasta din urma este mai **neteda**:

![ref2](C0vsC1.png 'Functia lui Weierstrass comparata cu una dintre primitivele sale')

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Nu putem obtine un $S$ de clasa $C^1$ cu polinoame de imbinare de grad 1, ci este nevoie ca functiile de imbinare sa fie **polinoame de grad 3 (functii cubice)**, intrucat cele de grad 2 ce ar rezulta in urma integrarii functiilor de imbinare de mai sus nu sunt destul de flexibile in practica. Pentru usurinta in calcule, vom considera:

$$s_i(x) = a_i(x - x_i)^3 + b_i(x - x_i)^2 + c_i(x - x_i) + d_i; \  a_i, b_i, c_i, d_i \in \mathbb{R}, i = \overline{0,n - 1}$$

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Acum avem $4n$ necunoscute, deci vom avea nevoie de tot atatea ecuatii. Daca am folosi doar conditiile de interpolare si de racordare de la Spline-urile de clasa $C^0$, la care am adauga conditia ca $S'$ sa fie continua, am avea doar $3n - 1$ ecuatii. Din acest motiv, se impune sa cunoastem si **derivata functiei date** in punctele-suport:

### Conditii de interpolare

$s_i(x_i) = f(x_i), \ i = \overline{0,n - 1} \Rightarrow n$ ecuatii

$s'_i(x_i) = f'(x_i), \ i = \overline{0,n - 1} \Rightarrow n$ ecuatii (aceasta conditie devine necesara din cauza numarului redus de ecuatii)

$s_{n - 1}(x_n) = f(x_n)$

$s'_{n - 1}(x_n) = f'(x_n)$

### Conditii de racordare

$s_i(x_{i + 1}) = s_{i + 1}(x_{i + 1}), \  i = \overline{0,n - 1}  \Rightarrow n - 1$ ecuatii

$s'_i(x_{i + 1}) = s'_{i + 1}(x_{i + 1}), \  i = \overline{0,n - 1}  \Rightarrow n - 1$ ecuatii

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Acum am obtinut cele $4n$ ecuatii necesare.

##  <div style="text-align: center"> 3. Interpolare cu functii de clasa $C^2$

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Dupa cum s-a vazut, marind clasa unei functii, aceasta devine mai neteda, iar in general, ceea ce se gaseste in natura, deci si datele care vor trebui la un moment dat interpolate sunt modelate dupa functii netede. Din acest motiv, se impune trecerea la interpolare cu functii Spline de clasa $C^2$, astfel incat si **curbura functiilor de imbinare** sa fie avuta in vedere. Pentru informatii aditionale cu privire la comportamentul unei curbe in functie de a doua derivata, a se vedea [Anexa](#anexa).

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Utilizarea unor functii Spline de clasa $C^2$ ofera un alt avantaj: acum avem suficiente ecuatii pentru a **nu mai avea nevoie de toate valorile derivatei functiei date**:

### Conditii de interpolare

$s_i(x_i) = f(x_i), \ i = \overline{0,n - 1} \Rightarrow n$ ecuatii

$s_{n - 1}(x_n) = f(x_n)$

### Conditii de racordare

$s_i(x_{i + 1}) = s_{i + 1}(x_{i + 1}), \  i = \overline{0,n - 1}  \Rightarrow n - 1$ ecuatii

$s'_i(x_{i + 1}) = s'_{i + 1}(x_{i + 1}), \  i = \overline{0,n - 1}  \Rightarrow n - 1$ ecuatii

$s''_i(x_{i + 1}) = s''_{i + 1}(x_{i + 1}), \  i = \overline{0,n - 1}  \Rightarrow n - 1$ ecuatii

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; In total $\textbf{4n - 2}$ **ecuatii**, dar ne trebuie $4n$ ecuatii. Din acest motiv, se introduc 2 tipuri de Spline-uri de clasa $C^2$:

### Spline-uri naturale

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Poarta denumirea aceasta deoarece la capetele setului de puncte date, graficul lui $S$ va fi drept, acest lucru datorandu-se alegerii:

$$s''(x_0) = 0 \textrm{ si } s''_{n - 1}(x_n) = 0$$

### Spline-uri tensionate

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Daca este posibil, se prefera sa se lucreze la capetele domeniului de interpolare cu derivatele functiei date, pentru a se utiliza mai multa informatie despre aceasta. Din acest motiv, Spline-urile tensionate sunt in general mai precise decat cele naturale:

$$s'(x_0) = f'(x_0) \textrm{ si } s'_{n - 1}(x_n) = f'(x_n)$$

# <div style="text-align: center"> Anexa - importanta dervatei a doua </div>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Mai sus s-a mentionat ca derivata a doua este o masura a **curburii** unei functii, pe care o vom nota cu $k$. Aceasta reprezinta cat de "stransa" este curba pe care o face graficul functiei intr-un anumit punct.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Mai exact, fie o functie $f$. Pentru orice punct $P(x, f(x)) \in G_f$, $\exists!$ un cerc de raza $r$ ce aproximeaza curba intr-o vecinatate a lui $P$. Curbura in $P$ este definita ca fiind:

$$k_x = \frac{1}{r} = \frac{|f''(x)|}{(1 + f'^2(x))^{\frac{3}{2}}}$$
